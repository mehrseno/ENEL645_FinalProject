```python
import pandas as pd
import random
import nltk
from nltk.corpus import wordnet

# Ensure that NLTK's resources are downloaded
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Your counts for each class
syn_counts = {
    'ENTP': 100, 'ENFP': 100,
    'ISFP': 100, 'ISTP': 100,
    'ENTJ': 100, 'ISTJ': 100, 'ENFJ': 100, 'ISFJ': 100,
    'ESTP': 100, 'ESFP': 100, 'ESFJ': 100, 'ESTJ': 100
}


# Function to get synonyms from WordNet
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            if lemma.name() != word and '_' not in lemma.name():
                synonyms.add(lemma.name().replace('_', ' '))
    return list(synonyms)

# Function for synonym replacement
def synonym_replacement(text, n):
    words = text.split()
    new_words = words.copy()
    random_words = list(set([word for word in words if wordnet.synsets(word)]))
    random.shuffle(random_words)
    num_replaced = 0
    for random_word in random_words:
        synonyms = get_synonyms(random_word)
        if synonyms:
            synonym = random.choice(synonyms)
            new_words = [synonym if word == random_word else word for word in new_words]
            num_replaced += 1
        if num_replaced >= n:  # only replace up to n words
            break
    
    sentence = ' '.join(new_words)
    return sentence

augmented_data = []

# Generate new texts with synonym replacement
for class_type, num_new_texts in syn_counts.items():
    class_rows = X_train[X_train['type'] == class_type]
    # Sample the number of rows required
    sampled_rows = class_rows.sample(n=num_new_texts, replace=True) # Set replace=True to allow duplicates
    for _, row in sampled_rows.iterrows():
        n_replace = max(1, len(row['Preprocessed_posts'].split()) // 40)  # Decide the number of words to replace
        augmented_text = synonym_replacement(row['Preprocessed_posts'], n_replace)
        augmented_data.append({
            'original_text': row['Preprocessed_posts'],  # original text
            'Preprocessed_posts': augmented_text,  # synonym replaced text
            'type': row['type']
        })

# Convert augmented data to DataFrame
synonym_df = pd.DataFrame(augmented_data)

# Save the augmented dataset
synonym_df.to_csv('Dataset_synonym_replacement_Final.csv', index=False)
```

# BERT Embedding

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Retrieve the Hugging Face API token from an environment variable
huggingface_token = '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'
# Set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Specify your custom cache directory again
custom_cache_dir = '/home/mehrnaz.senobarivayg/cache/'

# Initialize Longformer tokenizer and model with the token
tokenizer = AutoTokenizer.from_pretrained('google-bert/bert-large-uncased', cache_dir=custom_cache_dir, use_auth_token=huggingface_token)
model_bert = AutoModel.from_pretrained('google-bert/bert-large-uncased', cache_dir=custom_cache_dir, use_auth_token=huggingface_token, output_hidden_states=True).to(device)

# Set the model to evaluation mode
model_bert.to(device).eval()
```

```python

MAX_LEN = 512  # Set the maximum length
BATCH_SIZE = 8  # Set the batch size
```

```python


# Function to tokenize sentences
def tokenize_sentences(sentences, tokenizer, max_seq_len):
    tokenized_sentences = []
    for sentence in sentences:
        tokenized_sentence = tokenizer.encode(
            sentence,                       # Sentence to encode
            add_special_tokens=True,        # Add '[CLS]' and '[SEP]'
            max_length=max_seq_len,         # Truncate all sentences
            truncation=True                 # Explicitly truncate examples to max length
        )
        tokenized_sentences.append(tokenized_sentence)
    return tokenized_sentences



# Function to create attention masks
def create_attention_masks(tokenized_and_padded_sentences):
    attention_masks = []
    for sentence in tokenized_and_padded_sentences:
        att_mask = [int(token_id > 0) for token_id in sentence]
        attention_masks.append(att_mask)
    return np.asarray(attention_masks)
    
    
# Function to get BERT embeddings with batching
def get_bert_embeddings(sentences, tokenizer, model_bert, max_seq_len, batch_size, layer):
    # Tokenize and pad sentences
    input_ids = tokenize_sentences(sentences, tokenizer, max_seq_len)
    input_ids = pad_sequences(input_ids, maxlen=max_seq_len, dtype="long", value=0, truncating="post", padding="post")
    attention_masks = create_attention_masks(input_ids)

    # Convert inputs to PyTorch tensors
    input_ids = torch.tensor(input_ids).to(device)
    attention_masks = torch.tensor(attention_masks).to(device)

    cls_out_list = []
    hidden_states_list = []

    # Loop through batches
    for i in tqdm(range(0, input_ids.size(0), batch_size)):
        batch_input_ids = input_ids[i:i+batch_size]
        batch_attention_masks = attention_masks[i:i+batch_size]

        with torch.no_grad():
            outputs = model_bert(batch_input_ids, 
                                 attention_mask=batch_attention_masks,
                                 output_hidden_states=True)

        cls_tokens = outputs.last_hidden_state[:, 0, :].detach().cpu().numpy()
        hidden_states = outputs.hidden_states[-layer].detach().cpu().numpy()

        cls_out_list.append(cls_tokens)
        hidden_states_list.append(hidden_states)

    # Concatenate the results
    cls_out = np.vstack(cls_out_list)
    hidden_states_concat = np.vstack(hidden_states_list)

    return cls_out, hidden_states_concat
```
